```python
# Q1: What is statistics, and why is it important?
# A1: Statistics is the science of collecting, analyzing, interpreting, and presenting data.
#     It is important because it helps in decision-making based on data insights.

# Q2: What are the two main types of statistics?
# A2: Descriptive statistics and Inferential statistics.

# Q3: What are descriptive statistics?
# A3: Descriptive statistics summarize and describe the features of a dataset (e.g., mean, median, mode, standard deviation).

# Q4: What is inferential statistics?
# A4: Inferential statistics use sample data to make generalizations or predictions about a population.

# Q5: What is sampling in statistics?
# A5: Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population.

# Q6: What are the different types of sampling methods?
# A6: Types include:
#     1. Simple Random Sampling
#     2. Systematic Sampling
#     3. Stratified Sampling
#     4. Cluster Sampling
#     5. Convenience Sampling

# Q7: What is the difference between random and non-random sampling?
# A7: Random sampling ensures every member has an equal chance of selection; non-random sampling does not, which may introduce bias.

# Q8: Define and give examples of qualitative and quantitative data
# A8: Qualitative (categorical): Colors, gender, nationality.
#     Quantitative (numerical): Age, salary, height.

# Q9: What are the different types of data in statistics?
# A9: Types include:
#     1. Nominal
#     2. Ordinal
#     3. Interval
#     4. Ratio

# Q10: Explain nominal, ordinal, interval, and ratio levels of measurement
# A10:
#   - Nominal: Categories with no order (e.g., gender)
#   - Ordinal: Categories with order (e.g., education level)
#   - Interval: Ordered, equal intervals, no true zero (e.g., temperature in Celsius)
#   - Ratio: Ordered, equal intervals, true zero (e.g., weight, height)

# Q11: What is the measure of central tendency?
# A11: It refers to values that represent the center or average of a dataset (mean, median, mode).

# Q12: Define mean, median, and mode
# A12:
#   - Mean: Average of all values
#   - Median: Middle value when sorted
#   - Mode: Most frequent value

# Q13: What is the significance of the measure of central tendency?
# A13: It provides a single value that summarizes an entire dataset and helps understand the general trend.

# Q14: What is variance, and how is it calculated?
# A14: Variance measures the average squared deviation from the mean.
#      Formula: Var(X) = Î£(x - mean)Â² / N

# Q15: What is standard deviation, and why is it important?
# A15: Standard deviation is the square root of variance. It indicates how spread out the values are around the mean.

# Q16: Define and explain the term range in statistics
# A16: Range is the difference between the maximum and minimum values in a dataset.

# Q17: What is the difference between variance and standard deviation?
# A17: Variance is the average squared deviation; standard deviation is the square root of variance.
#      SD is in the same unit as the data, variance is not.

# Q18: What is skewness in a dataset?
# A18: Skewness measures the asymmetry of the distribution of values.

# Q19: What does it mean if a dataset is positively or negatively skewed?
# A19:
#   - Positive skew: Tail on the right, mean > median
#   - Negative skew: Tail on the left, mean < median

# Q20: Define and explain kurtosis
# A20: Kurtosis measures the "tailedness" of the distribution.
#      High kurtosis = heavy tails; low kurtosis = light tails.

# Q21: What is the purpose of covariance?
# A21: Covariance indicates the direction of the linear relationship between two variables.

# Q22: What does correlation measure in statistics?
# A22: Correlation measures the strength and direction of the linear relationship between variables (ranges from -1 to 1).

# Q23: What is the difference between covariance and correlation?
# A23:
#   - Covariance indicates direction but not strength (unit-dependent)
#   - Correlation is scaled (unitless), and shows both direction and strength

# Q24: What are some real-world applications of statistics?
# A24: Applications include:
#   - Healthcare (clinical trials)
#   - Business (market analysis)
#   - Sports (performance metrics)
#   - Government (census)
#   - Education (test scoring and analysis)
```

```python
#12
import pandas as pd
import numpy as np

# Create dataset
population = pd.DataFrame({'ID': range(1, 101), 'Value': np.random.randint(10, 100, 100)})

# Simple Random Sampling (sample 10)
simple_sample = population.sample(n=10, random_state=42)

# Systematic Sampling (every 10th element)
k = 10
start = np.random.randint(0, k)
systematic_sample = population.iloc[start::k][:10]

print("Simple Random Sampling:\n", simple_sample)
print("\nSystematic Sampling:\n", systematic_sample)
```

```python
#13

import pandas as pd

# Raw data with class labels
data = pd.DataFrame({
    'Score': [12, 15, 18, 22, 27, 33, 36, 38, 40],
    'Group': ['10-20', '10-20', '10-20', '20-30', '20-30', '30-40', '30-40', '30-40', '30-40']
})

# Group by class intervals and calculate mean
grouped_mean = data.groupby('Group')['Score'].mean()
print("Mean per group:\n", grouped_mean)

grouped_mode = data.groupby('Group')['Score'].apply(lambda x: x.mode())
print("Mode per group:\n", grouped_mode)

grouped_med = data.groupby('Group')['Score'].median()
print("Mean per group:\n", grouped_med)
```

```python
#14

import numpy as np
import pandas as pd
from scipy import stats

# 1. Simulate data (e.g., exam scores)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=100)  # mean=70, std=10, n=100

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Score'])

# 2. Central Tendency
mean = df['Score'].mean()
median = df['Score'].median()
mode = df['Score'].mode().values.tolist()

# 3. Dispersion
data_range = df['Score'].max() - df['Score'].min()
variance = df['Score'].var()
std_dev = df['Score'].std()

# 4. Display results
print("Central Tendency:")
print(f"  Mean   : {mean:.2f}")
print(f"  Median : {median:.2f}")
print(f"  Mode   : {mode}")

print("\nDispersion:")
print(f"  Range      : {data_range:.2f}")
print(f"  Variance   : {variance:.2f}")
print(f"  Std Dev    : {std_dev:.2f}")
```
