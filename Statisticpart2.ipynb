```python
# 1 What is hypothesis testing in statistics?
# Hypothesis testing is a statistical method used to make decisions about population parameters
# based on sample data. It evaluates whether there is enough evidence in a sample to support a specific claim.

# 2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
# The null hypothesis (H0) is a statement that there is no effect or no difference.
# The alternative hypothesis (H1 or Ha) is what you want to prove — it suggests there is an effect or a difference.

# 3 What is the significance level in hypothesis testing, and why is it important?
# The significance level (alpha, typically 0.05) is the threshold for rejecting the null hypothesis.
# It represents the probability of making a Type I error — falsely rejecting a true null hypothesis.

# 4 What does a P-value represent in hypothesis testing?
# The P-value is the probability of obtaining a test statistic at least as extreme as the one observed,
# assuming the null hypothesis is true.

# 5 How do you interpret the P-value in hypothesis testing?
# - If P-value ≤ alpha: reject H0 (evidence supports H1)
# - If P-value > alpha: fail to reject H0 (insufficient evidence to support H1)

# 6 What are Type 1 and Type 2 errors in hypothesis testing?
# - Type I Error: Rejecting H0 when it is actually true (false positive)
# - Type II Error: Failing to reject H0 when H1 is actually true (false negative)

# 7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
# - One-tailed test checks for deviation in one direction (e.g., > or <).
# - Two-tailed test checks for deviation in both directions (≠).

# 8 What is the Z-test, and when is it used in hypothesis testing?
# A Z-test is used to compare sample and population means when the population variance is known
# and the sample size is large (n ≥ 30).

# 9 How do you calculate the Z-score, and what does it represent in hypothesis testing?
# Z = (X̄ - μ) / (σ / √n)
# It measures how many standard deviations the sample mean (X̄) is from the population mean (μ).

# 10 What is the T-distribution, and when should it be used instead of the normal distribution?
# The T-distribution is used when the sample size is small (n < 30) and/or the population standard deviation is unknown.
# It has heavier tails than the normal distribution.

# 11 What is the difference between a Z-test and a T-test?
# - Z-test: Used when population variance is known and sample size is large.
# - T-test: Used when population variance is unknown and sample size is small.

# 12 What is the T-test, and how is it used in hypothesis testing?
# The T-test compares sample means to determine if they are significantly different.
# It's used when population variance is unknown.

# 13 What is the relationship between Z-test and T-test in hypothesis testing?
# Both test hypotheses about means, but Z-test uses known population standard deviation,
# while T-test uses sample standard deviation, making it more appropriate for small samples.

# 14 What is a confidence interval, and how is it used to interpret statistical results?
# A confidence interval is a range around a sample estimate that likely contains the true population parameter.
# It helps quantify the uncertainty of an estimate (e.g., 95% confidence means if we repeat the experiment 100 times,
# 95 intervals will contain the true mean).

# 15 What is the margin of error, and how does it affect the confidence interval?
# The margin of error defines the range above and below the sample estimate in the confidence interval.
# Larger margin of error → wider interval → more uncertainty.

# 16 How is Bayes' Theorem used in statistics, and what is its significance?
# Bayes' Theorem calculates the probability of an event based on prior knowledge of related conditions.
# It is fundamental in Bayesian statistics, updating beliefs as more data becomes available.

# 17 What is the Chi-square distribution, and when is it used?
# The Chi-square distribution is used in categorical data analysis, such as testing for independence
# or goodness of fit. It is skewed right and only takes positive values.

# 18 What is the Chi-square goodness of fit test, and how is it applied?
# It tests whether a categorical dataset matches an expected distribution.
# Used when you want to see if observed frequencies differ significantly from expected frequencies.

# 19 What is the F-distribution, and when is it used in hypothesis testing?
# The F-distribution arises when comparing two variances. It's used in ANOVA and regression to test significance.

# 20 What is an ANOVA test, and what are its assumptions?
# ANOVA (Analysis of Variance) tests whether there are statistically significant differences between the means of three or more groups.
# Assumptions: independence, normality, and homogeneity of variances.

# 21 What are the different types of ANOVA tests?
# - One-way ANOVA: One independent variable with multiple groups.
# - Two-way ANOVA: Two independent variables.
# - Repeated measures ANOVA: Same subjects used for each treatment.

# 22 What is the F-test, and how does it relate to hypothesis testing?
# The F-test compares two variances and is used in ANOVA and regression models to determine if model terms are significant.
```

```python
#3
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)

x_vals = [0, 1]
pdf_vals = [bernoulli_pdf(0.6, x) for x in x_vals]
plt.figure()
plt.bar(x_vals, pdf_vals)
plt.title("Bernoulli Distribution (p=0.6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
```

```python
#9
def compute_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Example usage with normal data
z_scores = compute_z_scores(data)

# Plot Z-scores
plt.hist(z_scores, bins=30, edgecolor='black', color='lightgreen')
plt.title("Z-Scores of Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#10
# Simulate non-normal (exponential) distribution
original_data = np.random.exponential(scale=2.0, size=10000)

# Parameters
sample_size = 50
n_samples = 1000
sample_means = []

# Generate sample means
for _ in range(n_samples):
    sample = np.random.choice(original_data, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot original distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(original_data, bins=50, edgecolor='black', color='orange')
plt.title("Original Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot distribution of sample means (CLT)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, edgecolor='black', color='deepskyblue')
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
