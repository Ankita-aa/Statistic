```python
# 1 What is a random variable in probability theory?
# A random variable is a numerical outcome of a random phenomenon.
# It maps outcomes of a random process to real numbers.

# 2 What are the types of random variables?
# There are two types: Discrete (countable values) and Continuous (any value in a range).

# 3 What is the difference between discrete and continuous distributions?
# Discrete: Deals with countable values (e.g., number of heads in coin toss).
# Continuous: Deals with uncountable, infinite values within a range (e.g., height, weight).

# 4 What are probability distribution functions (PDF)?
# PDF defines the probability of a continuous random variable falling within a range of values.
# It describes the density, not the exact probability at a point (which is 0 for continuous vars).

# 5 How do cumulative distribution functions (CDF) differ from PDFs?
# CDF gives the probability that a variable is less than or equal to a certain value.
# It is the integral (cumulative sum) of the PDF.

# 6 What is a discrete uniform distribution?
# A distribution where all outcomes have equal probability.
# E.g., rolling a fair die: P(1) = P(2) = ... = P(6) = 1/6.

# 7 What are the key properties of a Bernoulli distribution?
# Has two outcomes: 1 (success) with probability p, and 0 (failure) with probability 1-p.
# Mean = p, Variance = p(1 - p)

# 8 What is the binomial distribution, and how is it used?
# Models the number of successes in n independent Bernoulli trials.
# Parameters: n (trials), p (probability of success).
# Used to model binary events like pass/fail, win/lose.

# 9 What is the Poisson distribution and where is it applied?
# Models number of events in fixed interval (time, space) with known constant mean rate λ.
# Used in call center traffic, number of decay events, etc.

# 10 What is a continuous uniform distribution?
# Every value within a given interval [a, b] has equal likelihood.
# PDF = 1 / (b - a) for x in [a, b]

# 11 What are the characteristics of a normal distribution?
# Bell-shaped, symmetric around mean μ, defined by mean and std deviation σ.
# Most real-world phenomena approximate this (e.g., height, test scores).

# 12 What is the standard normal distribution, and why is it important?
# A normal distribution with mean = 0 and std deviation = 1.
# It's used for standardizing values (Z-scores) and in Z-tests.

# 13 What is the Central Limit Theorem (CLT), and why is it critical?
# CLT states that the sampling distribution of the sample mean approaches normality
# as sample size grows, regardless of the population distribution.
# Crucial for making inferences using normal approximation.

# 14 How does CLT relate to the normal distribution?
# It justifies the use of normal distribution for inference on means from any population,
# especially with large samples (n ≥ 30).

# 15 What is the application of Z statistics in hypothesis testing?
# Z-statistics standardizes the test statistic to compare against the standard normal distribution.
# Used when population variance is known.

# 16 How do you calculate a Z-score, and what does it represent?
# Z = (X - μ) / σ
# It shows how many standard deviations a value X is from the mean.

# 17 What are point estimates and interval estimates?
# Point estimate: single value estimate (e.g., sample mean) for a parameter.
# Interval estimate: range of values (e.g., confidence interval) believed to contain the parameter.

# 18 What is the significance of confidence intervals?
# Confidence interval gives a range of plausible values for a population parameter.
# E.g., 95% CI means we are 95% confident the true mean lies within the interval.

# 19 What is the relationship between a Z-score and a confidence interval?
# Z-scores determine the margin of error for the confidence interval.
# E.g., for 95% CI, Z ≈ 1.96 → CI = X̄ ± 1.96 * (σ / √n)

# 20 How are Z-scores used to compare different distributions?
# Z-scores normalize values, allowing comparisons across different scales/distributions.

# 21 What are the assumptions for applying the CLT?
# 1. Independent samples
# 2. Identically distributed population
# 3. Sufficiently large sample size (n ≥ 30 usually)

# 22 What is the concept of expected value in a probability distribution?
# Expected value (E[X]) is the mean of the distribution; the average outcome if experiment repeated infinitely.

# 23 How does a probability distribution relate to expected outcome?
# It defines how likely each outcome is, and the expected value is the weighted average using these probabilities.
```
